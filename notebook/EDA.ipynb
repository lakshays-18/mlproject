## Creating new columns Total Score and average

```python
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['average'] = df['total_score']/3
```

```python
df
```

## Data Visualization

```python
plt.figure(figsize=(8,8))
sns.histplot(data=df,x='average',kde=True,bins=10,hue='gender')
plt.show()
```

#### Insights
- Female Students have scored more than male students

## Count of how many have scored full marks and less than 20 marks and visualizing it in pie chart

```python
count_top_marks = []
count_least_marks = []

cols = ['math score','writing score','reading score']
for col in cols:
    count = int(df[df[col]==100]['average'].count())
    count_top_marks.append(count)
    count_less = int(df[df[col]<=20]['average'].count())
    count_least_marks.append(count_less)
```

```python
print(count_top_marks)
```

```python
plt.figure(figsize=(5,5))
plt.pie(count_top_marks,labels=['math','writing','reading'],autopct=lambda p: f'{int(p * sum(count_top_marks) / 100)}')
plt.title('Number of students who have scored 100 in respective subjects')
plt.figure(figsize=(5,5))
plt.pie(count_least_marks,labels=['math','writing','reading'],autopct=lambda p: f'{int(p * sum(count_least_marks) / 100)}')
plt.title('Number of students who have scored less than 20 in respective subjects')
plt.show()
```

Insights 
- Students scored full marks in:
   - Maths : 7
   - Writing : 14
   - Reading : 17

- Students scored less than < 20 marks in:
   - Maths : 4
   - Writing : 3
   - Reading : 1
